# Part II: Bayesian Inference

Is there a statistically significant difference between terror-related American deaths in the 15 years before 9/11 vs. the 15 years after?

```python
df[(df.country_txt == 'United States') & (~df.addnotes.str.contains('This attack was one of four related incidents',
                                                                    na=False))].groupby('iyear').nkillus.sum().plot(
    kind='bar', title='American Citizens killed by terror attacks in the US, Excluding 9/11', fontsize=9)
```

```python
deaths85to01 = df[(df.country_txt == 'United States') &
                  (~df.addnotes.str.contains('This attack was one of four related incidents', na=False)) &
                  (df.iyear > 1984) & (df.iyear < 2002)].groupby('iyear').nkillus.sum().values

deaths02to15 = df[(df.country_txt == 'United States') &
                  (~df.addnotes.str.contains('This attack was one of four related incidents', na=False)) &
                  (df.iyear > 2001)].groupby('iyear').nkillus.sum().values

mean_prior_mean = df[(df.country_txt == 'United States') &
                  (~df.addnotes.str.contains('This attack was one of four related incidents', na=False)) &
                  (df.iyear < 1985)].groupby('iyear').nkillus.sum().mean()

mean_prior_std = df[(df.country_txt == 'United States') &
                  (~df.addnotes.str.contains('This attack was one of four related incidents', na=False)) &
                  (df.iyear < 1985)].groupby('iyear').nkillus.sum().std(ddof=1)
```

```python
with pm.Model() as model:

    group85to01_mean = pm.Normal('deaths_85_01_mean', mean_prior_mean, sd=mean_prior_std)
    group02to15_mean = pm.Normal('deaths_02_15_mean', mean_prior_mean, sd=mean_prior_std)
    
std_prior_lower = 0.01
std_prior_upper = 100.0
    
with model:
    
    group85to01_std = pm.Uniform('deaths_85_01_std', lower=std_prior_lower, upper=std_prior_upper)
    group02to15_std = pm.Uniform('deaths_02_15_std', lower=std_prior_lower, upper=std_prior_upper)
```

```python
with model:

    group85to01 = pm.Normal('deaths_85_01', mu=group85to01_mean, sd=group85to01_std, observed=deaths85to01)
    group02to15 = pm.Normal('deaths_02_15', mu=group02to15_mean, sd=group02to15_std, observed=deaths02to15)
```

```python
with model:

    diff_of_means = pm.Deterministic('difference of means', group85to01_mean - group02to15_mean)
    diff_of_stds = pm.Deterministic('difference of stds', group85to01_std - group02to15_std)
    effect_size = pm.Deterministic('effect size',
                                   diff_of_means / np.sqrt((group85to01_std**2 + group02to15_std**2) / 2))
```

```python
with model:
    trace = pm.sample(25000, njobs=-1)
```

```python
pm.plot_posterior(trace[3000:],
                  varnames=['deaths_85_01_mean', 'deaths_02_15_mean', 'deaths_85_01_std', 'deaths_02_15_std'],
                  color='#87ceeb')
```

```python
pm.plot_posterior(trace[3000:],
                  varnames=['difference of means', 'difference of stds', 'effect size'],
                  ref_val=0,
                  color='#87ceeb')
```
